```python
import statistics

class InventoryItem:
    def __init__(self, name, stock, price_per_unit):
        self.name = name
        self.stock = stock
        self.price_per_unit = price_per_unit



    def sell(self,units):
        self.stock -= units

    def restock(self, units):
        self.stock += units

    def stock_value(self):
        stockval = self.stock * self.price_per_unit
        return stockval

    def needs_reorder(self,reorder_point):
        self.stock <= reorder_point
            

class Warehouse:
    def __init__(self,name):
        self.name = name
        self.items = {}


    def add_item(self,item):
        self.items[item.name] = item

    def transfer_to(self,other_warehouse, product_name, qty):
        if product_name in self.items and product_name in other_warehouse.items:
            self.items[product_name].sell(qty)
            other_warehouse.items[product_name].restock(qty)
            print(f"Transferred {qty} units of '{product_name}' from {self.name} to {other_warehouse.name}")
        else:
            print("Transfer failed: Item not found in one of the warehouses.")
            
    def low_stock_report(self,reorder_point):
        low_stock_items = [name, item in self.items if item.needs_reorder(reorder_point)]
        
# Create items
p1 = InventoryItem("Laptop", 150, 60000)
p2 = InventoryItem("Mobile", 80, 30000)

# Create warehouses
w1 = Warehouse("Warehouse A")
w2 = Warehouse("Warehouse B")

# Add items
w1.add_item(p1)
w2.add_item(p2)

# Transfer
w1.transfer_to(w2, "Laptop", 20)

# Low stock report
print("Low stock items in Warehouse A:", w1.low_stock_report(100))




    
```
