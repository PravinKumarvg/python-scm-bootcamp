```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np


data = pd.read_csv("StockMovementdate.csv")
print(data.head())
print(data.tail())
print(data.isna())


# Convert column to datetime
data["Last_Movement_Date"] = pd.to_datetime(data["Last_Movement_Date"])

# Get current date
current_date = pd.Timestamp.today()

# Calculate difference in days
data["date_diff"] = (current_date - data["Last_Movement_Date"]).dt.days

# Categorize based on difference


data["Category"] = np.where(data["date_diff_days"] < 30, "Healthy",
                    np.where(data["date_diff_days"] < 90, "Slow Moving", "Dead Stock"))

data.head()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np


data = pd.read_csv("Safetystockrep.csv")
print(data.head())
print(data.tail())
print(data.isna())
print(data.columns)


data["SafetyStock"] = data["Z"] * data["Std_Demand"] * np.sqrt(data["LeadTime_Days"])
data["ReorderPoint"] = data["Mean_Demand"] * data["LeadTime_Days"] + data["SafetyStock"]

print(data.head())
```
