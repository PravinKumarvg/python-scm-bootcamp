```python
#Nested Functions

def add(n1, n2):
    return n1+n2

def sub(n1, n2):
    return n1-n2

def mul(n1, n2):
    return n1*n2


def calculate(calc_func, n1, n2):
    return calc_func(n1,n2)


result = calculate(add, 3,2)#We can pass function as we pass arguments
print(result)
```

```python
#Nested functions

def outer_function():
    print("I am outer")
    
    def nested_function():
        return "Hello from nested function"
    return nested_function  #This line is crucial

inner_function = outer_function()
inner_function() 
```

```python
#Decorator function

import time

def delay_func(function):
    def wrapper_fun():
        time.sleep(2)
        function()

    return wrapper_fun


@delay_func
def say():
    print("Say Hello")

@delay_func
def greet():
    print("Gud Mrng")

def thank():
    print("Thank you so much")


thank()
```

```python
#Decorator function

import time

# ðŸ‘‡ This is your decorator
def speed_calc_decorator(function):
    def wrapper_function():
        starttime = time.time()
        
        endtime = time.time()
        approxtime = endtime - starttime
        print(f"{function.__name__} run speed: {approxtime}s")
    return wrapper_function


# ðŸ‘‡ Fast function (runs quickly)
@speed_calc_decorator
def fast_function():
    for i in range(1000000):
        i * i

# ðŸ‘‡ Slow function (runs slowly)
@speed_calc_decorator
def slow_function():
    for i in range(10000000):
        i * i

# ðŸ‘‡ Run both
fast_function()
slow_function()
```
