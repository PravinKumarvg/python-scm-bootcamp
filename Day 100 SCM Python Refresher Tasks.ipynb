```python
#Class refresher task

class Product:
    def __init__(self, name, stock, price):
        self.name = name
        self.stock = stock
        self.price = price

    def sell(self, units):
        self.stock -= units
        print(f"The quantity {units} was sold")

    def restock(self, units):
        self.stock += units
        print(f"The quantity {units} was sold")

    def stockvalue(self):
        stockval = self.stock * self.price
        print(f"The stock value is {stockval}")


class WarehouseProduct(Product):
    def __init__(self, name, stock, price, warehousename):
        super().__init__(name, stock, price)
        self.warehousename = warehousename

    def transfer(self,qty, destination):
        self.stock -= qty
        print(f" The stock quantity {qty} is transferred to {destination}")

p1 = WarehouseProduct("Laptop", 200, 50000, "Chennai DC")
sel = p1.sell(20)
res = p1.restock(50)
p1.transfer(40, "Mumbai DC")




        
```
