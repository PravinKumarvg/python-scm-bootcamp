```python
#Using Selenium to import to Tender and automate responses

from selenium import webdriver
from selenium.webdriver.common.by import By
from time import sleep

# Replace with your actual credentials
FB_EMAIL = "your_facebook_email"
FB_PASSWORD = "your_facebook_password"

# Launch browser
driver = webdriver.Chrome()

# Open Tinder
driver.get("https://www.tinder.com")
sleep(2)

# Click "Log in" using link text
login_button = driver.find_element(By.LINK_TEXT, "Log in")
login_button.click()
sleep(2)

# Click "Log in with Facebook" using CSS selector
fb_login_button = driver.find_element(By.CSS_SELECTOR, "button[aria-label='Log in with Facebook']")
fb_login_button.click()
sleep(2)

# Switch to Facebook login popup
driver.switch_to.window(driver.window_handles[1])

# Enter Facebook credentials
email_input = driver.find_element(By.ID, "email")
email_input.send_keys(FB_EMAIL)

password_input = driver.find_element(By.ID, "pass")
password_input.send_keys(FB_PASSWORD)

login_submit = driver.find_element(By.NAME, "login")
login_submit.click()

# Switch back to Tinder window
driver.switch_to.window(driver.window_handles[0])       

print("Current page:", driver.title)

# Wait for page to load
sleep(5)

# Allow location using aria-label or button order
try:
    allow_location_button = driver.find_element(By.CSS_SELECTOR, "button[aria-label='Allow']")
    allow_location_button.click()
    print("✅ Location allowed.")
except:
    print("❌ Location button not found.")

# Disallow notifications using aria-label or button order
try:
    notifications_button = driver.find_element(By.CSS_SELECTOR, "button[aria-label='Not interested']")
    notifications_button.click()
    print("✅ Notifications disallowed.")
except:
    print("❌ Notifications button not found.")

# Accept cookies using class or button text
try:
    cookies_button = driver.find_element(By.CSS_SELECTOR, "button[class*='accept-button']")
    cookies_button.click()
    print("✅ Cookies accepted.")
except:
    print("❌ Cookies button not found.")


# Loop to like 100 profiles
for n in range(100):
    sleep(1)  # Add a 1-second delay between likes

    try:
        print("Like attempt:", n + 1)
        like_button = driver.find_element(By.CSS_SELECTOR, "button[aria-label='Like']")
        like_button.click()

    # If a match popup blocks the like button
    except ElementClickInterceptedException:
        try:
            match_popup = driver.find_element(By.CSS_SELECTOR, ".itsAMatch a")
            match_popup.click()
            print("Match popup closed.")
        except NoSuchElementException:
            print("Match popup not found. Waiting...")
            sleep(2)

# Close the browser after the loop
driver.quit()
```
