```python
import pandas as pd

# Read CSV files
data1 = pd.read_csv("Inv.csv")
data2 = pd.read_csv("suppprice.csv")

# Merge on 'Product' column using outer join
Totaldata = pd.merge(data1, data2, on="Product", how="outer")

# Calculate stock value
Totaldata["Stockvalue"] = Totaldata["Stock"] * Totaldata["UnitPrice"]

# Flag items below reorder point
Totaldata["Stock<Reorder"] = Totaldata["Stock"] < Totaldata["ReorderPoint"]
Totaldata["Stock<Reorder"] = Totaldata["Stock<Reorder"].apply(lambda x: "Yes" if x else "No")

# Group by Category
Total_stk = Totaldata.groupby("Category")

# Total stock value per category
Totalstockvalue = Total_stk["Stockvalue"].sum()

# Total stock quantity per category
Totalstock = Total_stk["Stock"].sum()

# Sort items by UnitPrice (descending)
lowstockitem = Totaldata.sort_values(by="UnitPrice", ascending=False)

# Print results
print("Total Stock Value by Category:\n", Totalstockvalue)
print("\nTotal Stock Quantity by Category:\n", Totalstock)
print("\nNumber of Items (sorted by UnitPrice):", len(lowstockitem))

# Export to CSV
Totaldata.to_csv("LowStockReport.csv", index=False)
```

```python
#numpy
import numpy as np

# Weekly demand data
weekly_demand = [100, 120, 110, 130, 125, 115, 140, 150, 135, 145, 160, 155]
lead_time_weeks = 3
safety_stock = 200
simulations = 1000

# Convert to NumPy array
demand_array = np.array(weekly_demand)

# Demand stats
mean = np.mean(demand_array)
std = np.std(demand_array)
cov = std / mean * 100

# Reorder quantity calculation
ReorderQty = (demand_array * lead_time_weeks) + safety_stock

# Min/Max demand week (Week #1â€“12)
min_week = np.argmin(demand_array) + 1
max_week = np.argmax(demand_array) + 1

# Monte Carlo simulation
total_demand_samples = []

for _ in range(simulations):
    random_weeks = np.random.normal(mean, std, lead_time_weeks)
    total_demand = np.sum(random_weeks)
    total_demand_samples.append(total_demand)

# Estimate stockout probability
stockouts = [x for x in total_demand_samples if x > safety_stock]
prob_stockout = len(stockouts) / simulations * 100

# Print results

print(f"COV (%): {cov:.2f}%")
print(f"Week with min demand: Week #{min_week}")
print(f"Week with max demand: Week #{max_week}")
print(f"Reorder Quantity:{ReorderQty}")
print(f"  Min: {np.min(ReorderQty)}")
print(f"  Mean: {np.mean(ReorderQty):.2f}")
print(f"  Max: {np.max(ReorderQty)}")
print(f"Estimated Stockout Probability: {prob_stockout:.2f}%")
```
