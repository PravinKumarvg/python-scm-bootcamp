```python
#Class Task

class InventoryItem:
    def __init__(self, name, stock, price_per_unit):
        self.name = name
        self.stock = stock
        self.price_per_unit = price_per_unit


    def sell(self,units):
        self.stock -= units
        return units
        
        
    def restock(self,units):
        self.stock += units
        return units
        
        
    def stock_value(self):
        stockvalue = self.stock * self.price_per_unit
        return stockvalue

    def reorderneeded(self, units):
        if self.stock < units:
            print("Reorder Needed")

        else:
            print("Stock is fine")


P1 = InventoryItem("Laptop", 400, 40)
P2 = InventoryItem("Mobile", 300, 30)
P3 = InventoryItem("Tablet", 200, 20)

P1sel = P1.sell(30)
P1res = P1.restock(60)
P1reord = P1.reorderneeded(50)

P2sel = P2.sell(280)
P2res= P2.restock(20)
P2reord = P2.reorderneeded(50)

P3sel= P3.sell(170)
P3res = P3.restock(20)
P3reord = P3.reorderneeded(50)



print(f"{P1.name} sold {P1sel} units and restocked {P1res} units. The stock value is {P1.stock_value()}")
print(f"{P2.name} sold {P2sel} units and restocked {P2res} units. The stock value is {P2.stock_value()}")
print(f"{P3.name} sold {P3sel} units and restocked {P3res} units. The stock value is {P3.stock_value()}")





```

```python
#function Reorder Quantity Calculator


def calculate_reorder_qty(daily_sales, lead_time, safety_stock):
    reorder_qty = daily_sales * lead_time + safety_stock

    
    if safety_stock < reorder_qty:
        print("Reorder qty Needed")

    else: 
        print("No reorder qty needed")

    return reorder_qty


daily_sale = int(input("What is daily sales?"))
Lt = int(input("What is leadtime ?"))
SS = int(input("What is the safety stock ?"))

reord = calculate_reorder_qty(daily_sale, Lt, SS)



    
```

```python
#Numpy
import numpy as np

demand = np.array([100,120,110,130,125,115,140,150,135,145,160,155])

mean = np.mean(demand)
std = np.std(demand)
cov = std/mean *100

HighestIndex = np.argmax(demand)
highest_sku = demand[HighestIndex]

LowestIndex = np.argmin(demand)
Lowest_sku = demand[LowestIndex]

print(f" The mean is {mean} and the standard deviation is {std}. The cov is {cov}")
print(f"Highest demand is {highest_sku}")
print(f"Lowest demand is {Lowest_sku}")
```

```python
#SCM Product List API

import requests

prodwebsite  = "https://fakestoreapi.com/products"

response = requests.get(prodwebsite)
response.raise_for_status
products = response.json()


product_list = [(product["title"], product["price"]) for product in products]

sorted_products = sorted(product_list, key=lambda x:x[1])

Print("Top 5 cheapest products")
for title ,price in sorted_products[:5]:
    print(f"{title}: {price}")

print("Eliginle for bulk order(price < 500)")

for title ,price in sorted_products:
    if price<500:
        print(f"{title}: {price}")
```
